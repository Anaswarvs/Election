**ABOUT THE DATASET**


**Content**

This database contains detailed candidatelevel data for elections to the lower houses of Indias national
and state legislatures, i.e., the Lok Sabha and Vidhan Sabhas. The data span 19772015, with each row
representing a candidate that ran for office in that stateyear.

The variables in the national election data file are:

**Variable name :       Variable content**

st_name :          State

Year :         General election year

pc_no :          Parliamentary constituency number

pc_name :          Parliamentary constituency name

pc_type :         Parliamentary constituency reservation status

cand_name :       Candidate name

cand_sex :       Candidate sex

partyname :       Party name

partyabbre :       Party abbreviation

totvotpoll :      Votes received

electors :       Number of registered voters

```python
#Encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['st_name']=le.fit_transform(df['st_name'])
df['pc_name']=le.fit_transform(df['pc_name'])
df['pc_type']=le.fit_transform(df['pc_type'])
df['cand_name']=le.fit_transform(df['cand_name'])
df['cand_sex']=le.fit_transform(df['cand_sex'])
df['partyname']=le.fit_transform(df['partyname'])
df['partyabbre']=le.fit_transform(df['partyabbre'])
```

```python
#outliers removing
for i in["year","pc_no","totvotpoll","electors"]:
 plt.figure(figsize=(6, 6))
 sns.boxplot(df[i], color='green')
 plt.xlabel(i)
 plt.ylabel('Range')
```

```python
def iqr_method(df,variables):
  q1=df[variables].quantile(0.25)
  q3=df[variables].quantile(0.75)
  iqr=q3-q1
  upper=q3+(1.5*iqr)
  lower=q1-(1.5*iqr)
  return lower,upper
```

```python
lower_lim,upper_lim=iqr_method(df,'totvotpoll')
print('lower limit = ',lower_lim)
print('upper limit = ',upper_lim )
```

```python
df['totvotpoll']=np.where(df['totvotpoll']>upper_lim,upper_lim,
                   np.where(df['totvotpoll']<lower_lim,lower_lim,df['totvotpoll']))
```

```python
lower_lim,upper_lim=iqr_method(df,'electors')
print('lower limit = ',lower_lim)
print('upper limit = ',upper_lim )
```

```python
df['electors']=np.where(df['electors']>upper_lim,upper_lim,
                   np.where(df['electors']<lower_lim,lower_lim,df['electors']))
```

```python
lower_lim,upper_lim=iqr_method(df,'pc_no')
print('lower limit = ',lower_lim)
print('upper limit = ',upper_lim )
```

```python
df['pc_no']=np.where(df['pc_no']>upper_lim,upper_lim,
                   np.where(df['pc_no']<lower_lim,lower_lim,df['pc_no']))
```

# **Model Creation**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_percentage_error,r2_score
lr=LinearRegression()
dectree=DecisionTreeRegressor()
rand=RandomForestRegressor(max_depth=25, random_state=100)
lst=[lr,dectree,rand]
```

```python
for i in lst:
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print("model is",i)
  print("percentage erreor is",mean_absolute_percentage_error(y_test,y_pred))
  print("r2score is",r2_score(y_test,y_pred))
  print()
```

```python
dic={'Model':["lr","dectree","rand"],'r2_score':[0.5893,0.9987,0.9980]}
result=pd.DataFrame(dic)
result
```
